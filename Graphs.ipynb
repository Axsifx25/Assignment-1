```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=[10,12]
y=[2,8]
plt.plot(x,y,marker='*' , color='blue', linestyle='-',linewidth=5,markersize=15)
plt.title('line plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

```python
x=[10,12]
y=[2,8]
plt.plot(x,y,marker='*' , color='blue', linestyle='-',linewidth=5,markersize=15)
plt.title('line plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()

plt.scatter(x,y, color='red',marker='x')
plt.title('Scatter plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
Name=['Umar','Asif']
age=[20,21]
plt.bar(Name,age,color='cyan')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
data=[2,3,4,7,8]
plt.hist(data,bins=10,color='teal',edgecolor='black')
plt.title('Histogram')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

```python
sizes= [20,45,80,51]
labels=['a','b','c','d']
plt.pie(sizes,labels=labels,autopct='%1.1f%%' , startangle=0)
plt.title('Pie Chart')
plt.show()
```
